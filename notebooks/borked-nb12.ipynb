## (c) Producing unit-area histograms

To produce a unit-area histogram, the `hist()` function is used again. However, this time we need to add the extra argument `prob = T` to the input values for the function to instruct R to produce a unit-area histogram instead of the default frequency histogram. This extra argument stands for 'probability = True' - that is, the area of each bar in the histogram is a probability, so the total area of the bars will equal 1. In other words, the bars will have 'unit area'. (Note that the `prob = T` argument is case sensitive.)

**(i)** Run the code below to produce a unit-area histogram of `strength`, and compare this histogram with the frequency histogram of `strength` produced in part (b)(i).

```r
hist(fifa19$strength, prob = T)
```

#### Solution
As you may have expected, the shape of the unit-area histogram is exactly the same as the shape of the frequency histogram. The only thing that is different is the vertical scale of the histogram.

**(ii)** Use the empty code cell below to produce unit-area histograms of `height`, `weight` and `marking`, and verify that the only difference between these unit-area histograms and their frequency histogram versions in part (b)(ii) is the vertical scale.

#### Solution
The required code is shown below. The resulting unit-area histograms are indeed the same as their frequency histogram versions except for the vertical scales.

```r
hist(fifa19$height, prob = T)
hist(fifa19$weight, prob = T)
hist(fifa19$marking, prob = T)
```
